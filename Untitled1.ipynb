```python
import numpy as np
from scipy.signal import spectrogram, csd, stft
import speech_recognition as sr
from scipy.io import wavfile
from matplotlib import pyplot as plt
import python_speech_features as psf
```

```python
recognizer = sr.Recognizer()
mic = sr.Microphone()

with mic as source:
    print("Adjusting mic...")
    recognizer.adjust_for_ambient_noise(mic, 3)
    print("Now you can talk")
    audio = recognizer.listen(mic, 3)
    print("Detecting what you said...")
```

```python
with open('audio.wav','wb') as f:
    f.write(audio.get_wav_data())
    f.close()
```

```python
sig = wavfile.read('audio.wav')

def inputize(signal):
    f, t, zxx = stft(signal[1], fs= signal[0],nperseg=1103)
    magspec = psf.base.sigproc.magspec(zxx, 1103)
    melspec = psf.hz2mel(magspec)
    log_mel = psf.logfbank(melspec, samplerate=44100, nfft=1103)
    inverted = np.fft.ifft(log_mel)
    
    return inverted.real
```

```python
result = inputize(sig)

plt.pcolormesh(result)
```
